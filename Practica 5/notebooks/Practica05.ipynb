## Objetivo General

Generar un dataset clínico **simulado** con información de **5,000 pacientes** del estado de Puebla para ser utilizado como base en prácticas de clasificación supervisada y análisis de riesgo cardiovascular.

> ⚠️ Los datos son completamente ficticios con fines académicos exclusivamente.

## 1. Importación de Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import warnings
import os

warnings.filterwarnings('ignore')
matplotlib.use('Agg')

# Configuracion visual
sns.set_theme(style='darkgrid', palette='husl')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.size'] = 11

# Crear directorio de salida si no existe
os.makedirs('../outputs', exist_ok=True)

print('Librerias cargadas correctamente.')
```

## 2. Carga del Dataset

**Tarea 12:** Cargar el dataset en Jupyter Notebook usando Pandas.

```python
# Carga del dataset desde la carpeta data/
ruta_csv = '../data/pacientes_puebla.csv'
df = pd.read_csv(ruta_csv)

# Asegurar tipos numericos correctos
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
df['CVE_MUN'] = df['CVE_MUN'].astype(str).str.zfill(3)

print('Dataset cargado exitosamente.')
print(f'Ruta: {ruta_csv}')
df.head(10)
```

## 3. Verificación de Dimensiones

**Tarea 13:** Confirmar que el dataset tenga 5,000 registros y el número correcto de columnas.

```python
filas, columnas = df.shape
print('=' * 50)
print(f'  Número de registros : {filas:,}')
print(f'  Número de columnas  : {columnas}')
print('=' * 50)

assert filas == 5000, f'ERROR: Se esperaban 5,000 registros, se encontraron {filas}'
assert columnas == 18, f'ERROR: Se esperaban 18 columnas, se encontraron {columnas}'
print('✔ Verificacion de dimensiones: CORRECTA')
```

## 4. Validación de Nombres de Columnas

**Tarea 14:** Revisar que las columnas tengan nombres claros, consistentes y sin espacios problemáticos.

```python
columnas_esperadas = [
    'ID_Paciente', 'Municipio', 'CVE_MUN', 'Latitude', 'Longitude',
    'Edad', 'Genero', 'Nivel_Socioeconomico', 'Acceso_Salud',
    'Presion_Sistolica', 'Presion_Diastolica', 'Colesterol_Total',
    'Glucosa', 'IMC', 'Fumador', 'Historial_Familiar',
    'Riesgo_Score', 'Diagnostico_Infarto'
]

print('Columnas encontradas:')
for i, col in enumerate(df.columns, 1):
    tiene_espacio = ' ' in col
    estado = '⚠ ESPACIO' if tiene_espacio else '✔'
    print(f'  {i:2d}. {col:<30} {estado}')

print(f'\nTotal columnas: {len(df.columns)}')
faltantes = set(columnas_esperadas) - set(df.columns)
if faltantes:
    print(f'Columnas faltantes: {faltantes}')
else:
    print('✔ Todas las columnas esperadas están presentes.')
```

## 5. Verificación de Tipos de Datos

**Tarea 15:** Comprobar que cada columna tenga el tipo de dato adecuado.

```python
print('Tipos de datos por columna:')
print('-' * 45)
for col, dtype in df.dtypes.items():
    print(f'  {col:<30} {str(dtype)}')

print('\nInformacion general del DataFrame:')
df.info()
```

## 6. Búsqueda de Valores Nulos

**Tarea 16:** Identificar si existen datos vacíos o incompletos.

```python
nulos = df.isnull().sum()
total_nulos = nulos.sum()

print('Valores nulos por columna:')
print(nulos.to_string())
print('-' * 40)
print(f'Total de valores nulos en el dataset: {total_nulos}')

if total_nulos == 0:
    print('✔ No se encontraron valores nulos. Dataset completo.')
else:
    print(f'⚠ Se encontraron {total_nulos} valores nulos. Requiere limpieza.')
```

## 7. Búsqueda de Registros Duplicados

**Tarea 17:** Verificar que no existan pacientes duplicados o registros repetidos.

```python
duplicados_total = df.duplicated().sum()
duplicados_id = df['ID_Paciente'].duplicated().sum()

print(f'Registros completamente duplicados : {duplicados_total}')
print(f'IDs de paciente duplicados          : {duplicados_id}')

if duplicados_total == 0 and duplicados_id == 0:
    print('✔ No se encontraron registros duplicados. Integridad correcta.')
else:
    print('⚠ Se encontraron duplicados. Requiere limpieza.')
```

## 8. Validación de Rangos Clínicos

**Tarea 18:** Revisar que los valores clínicos estén dentro de rangos lógicos y realistas.

```python
reglas = {
    'Edad':               (18, 90),
    'Presion_Sistolica':  (80, 200),
    'Presion_Diastolica': (50, 120),
    'Colesterol_Total':   (100, 300),
    'Glucosa':            (70, 300),
    'IMC':                (15.0, 45.0),
    'Riesgo_Score':       (0, 100),
}

print('Validacion de rangos clinicos:')
print('-' * 60)
todo_valido = True
for columna, (minimo, maximo) in reglas.items():
    fuera_rango = df[(df[columna] < minimo) | (df[columna] > maximo)].shape[0]
    estado = '✔' if fuera_rango == 0 else f'⚠ {fuera_rango} valores fuera de rango'
    print(f'  {columna:<25} [{minimo:>6} – {maximo:>6}]  {estado}')
    if fuera_rango > 0:
        todo_valido = False

print('-' * 60)
if todo_valido:
    print('✔ Todos los valores clinicos estan dentro de rangos validos.')
else:
    print('⚠ Se detectaron valores fuera de rango clinico.')
```

## 9. Validación de Datos Geográficos

**Tarea 19:** Confirmar que las coordenadas correspondan al estado de Puebla.

```python
# Limites geograficos del estado de Puebla
lat_min, lat_max = 17.80, 21.00
lon_min, lon_max = -99.10, -96.70

# Convertir coordenadas a numerico (por si fueron leidas como string)
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

fuera_lat = df[(df['Latitude'] < lat_min) | (df['Latitude'] > lat_max)].shape[0]
fuera_lon = df[(df['Longitude'] < lon_min) | (df['Longitude'] > lon_max)].shape[0]

print('Validacion de coordenadas geograficas (Estado de Puebla):')
print(f'  Latitud  valida [{lat_min} - {lat_max} N]  : {df.shape[0] - fuera_lat} correctas, {fuera_lat} fuera de rango')
print(f'  Longitud valida [{lon_min} - {lon_max} O] : {df.shape[0] - fuera_lon} correctas, {fuera_lon} fuera de rango')

print(f'\nMunicipios unicos en el dataset: {df["Municipio"].nunique()}')
print('\nTop 10 municipios con mas pacientes:')
print(df['Municipio'].value_counts().head(10).to_string())

if fuera_lat == 0 and fuera_lon == 0:
    print('\nTodas las coordenadas corresponden al estado de Puebla.')
else:
    print(f'\nSe encontraron {fuera_lat + fuera_lon} coordenadas fuera del estado de Puebla.')
```

## 10. Limpieza Básica de Datos

**Tarea 20:** Corregir errores detectados en nombres, tipos de datos, valores nulos, duplicados o rangos inválidos.

```python
df_limpio = df.copy()

# 1. Normalizar columnas categoricas (quitar espacios extra)
cols_cat = ['Genero', 'Nivel_Socioeconomico', 'Acceso_Salud', 'Fumador', 'Historial_Familiar']
for col in cols_cat:
    df_limpio[col] = df_limpio[col].astype(str).str.strip()

# 2. Convertir tipos de datos correctos
df_limpio['CVE_MUN'] = df_limpio['CVE_MUN'].astype(str).str.zfill(3)
df_limpio['Diagnostico_Infarto'] = df_limpio['Diagnostico_Infarto'].astype(int)

# 3. Eliminar duplicados si los hubiera
antes = len(df_limpio)
df_limpio = df_limpio.drop_duplicates()
despues = len(df_limpio)

print('Resumen del proceso de limpieza:')
print(f'  Registros antes de limpieza  : {antes:,}')
print(f'  Duplicados eliminados         : {antes - despues}')
print(f'  Registros despues de limpieza : {despues:,}')
print(f'  Valores nulos restantes       : {df_limpio.isnull().sum().sum()}')
print('\n✔ Limpieza basica completada exitosamente.')
```

## 11. Análisis Estadístico Inicial

**Tarea 21:** Obtener medidas básicas como promedio, mínimo, máximo y desviación estándar de las variables numéricas.

```python
cols_numericas = ['Edad', 'Presion_Sistolica', 'Presion_Diastolica',
                  'Colesterol_Total', 'Glucosa', 'IMC', 'Riesgo_Score']

estadisticas = df_limpio[cols_numericas].describe().T
estadisticas.columns = ['Conteo', 'Media', 'Desv_Std', 'Minimo', 'Q1', 'Mediana', 'Q3', 'Maximo']

print('Estadisticas descriptivas de variables numericas:')
print('=' * 80)
print(estadisticas.round(2).to_string())
print('=' * 80)
```

## 12. Visualizaciones EDA — Histogramas de Variables Numéricas

**Tarea 22:** Crear gráficas para explorar los datos.

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

colores = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']

for i, col in enumerate(cols_numericas):
    axes[i].hist(df_limpio[col], bins=30, color=colores[i], edgecolor='white', alpha=0.85)
    axes[i].set_title(f'Distribucion: {col}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frecuencia')
    axes[i].axvline(df_limpio[col].mean(), color='red', linestyle='--', linewidth=1.5, label='Media')
    axes[i].legend(fontsize=9)

axes[-1].set_visible(False)
fig.suptitle('Distribucion de Variables Numericas — Dataset Pacientes Puebla', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../outputs/histogramas_variables.png', dpi=150, bbox_inches='tight')
plt.show()
print('Grafica guardada en: ../outputs/histogramas_variables.png')
```

## 13. Visualizaciones EDA — Variables Categóricas

```python
cols_cat_plot = ['Genero', 'Nivel_Socioeconomico', 'Acceso_Salud', 'Fumador', 'Historial_Familiar']

fig, axes = plt.subplots(1, 5, figsize=(24, 6))

paletas = ['Set2', 'Set1', 'Paired', 'Accent', 'Dark2']

for i, col in enumerate(cols_cat_plot):
    conteo = df_limpio[col].value_counts()
    colores_bar = sns.color_palette(paletas[i], len(conteo))
    bars = axes[i].bar(conteo.index, conteo.values, color=colores_bar, edgecolor='white')
    axes[i].set_title(col, fontsize=12, fontweight='bold')
    axes[i].set_ylabel('Cantidad de Pacientes')
    for bar, val in zip(bars, conteo.values):
        axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
                     f'{val:,}', ha='center', va='bottom', fontsize=9)
    axes[i].tick_params(axis='x', rotation=20)

fig.suptitle('Distribucion de Variables Categoricas', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/variables_categoricas.png', dpi=150, bbox_inches='tight')
plt.show()
print('Grafica guardada en: ../outputs/variables_categoricas.png')
```

## 14. Análisis de la Distribución del Diagnóstico de Infarto

**Tarea 23:** Revisar cuántos pacientes pertenecen a cada clase.

```python
distribucion = df_limpio['Diagnostico_Infarto'].value_counts()
etiquetas = ['Sin Infarto (0)', 'Con Infarto (1)']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Grafica de pastel
colores_pie = ['#59a14f', '#e15759']
axes[0].pie(distribucion.values, labels=etiquetas, autopct='%1.1f%%',
            colors=colores_pie, startangle=90, textprops={'fontsize': 13},
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[0].set_title('Distribucion del Diagnostico de Infarto', fontsize=13, fontweight='bold')

# Grafica de barras
bars = axes[1].bar(etiquetas, distribucion.values, color=colores_pie, edgecolor='white', width=0.5)
for bar, val in zip(bars, distribucion.values):
    axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20,
                 f'{val:,} pacientes', ha='center', va='bottom', fontsize=12, fontweight='bold')
axes[1].set_title('Cantidad de Pacientes por Diagnostico', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Numero de Pacientes')
axes[1].set_ylim(0, max(distribucion.values) * 1.15)

fig.suptitle('Analisis de la Variable Objetivo: Diagnostico de Infarto', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/distribucion_diagnostico.png', dpi=150, bbox_inches='tight')
plt.show()

print('Resumen de distribucion:')
for clase, cantidad in distribucion.items():
    pct = cantidad / len(df_limpio) * 100
    etiqueta = 'Con Infarto' if clase == 1 else 'Sin Infarto'
    print(f'  {etiqueta} ({clase}): {cantidad:,} pacientes ({pct:.1f}%)')
```

## 15. Análisis por Nivel Socioeconómico y Diagnóstico

```python
nivel_diag = df_limpio.groupby(['Nivel_Socioeconomico', 'Diagnostico_Infarto']).size().unstack(fill_value=0)
nivel_diag.columns = ['Sin Infarto', 'Con Infarto']

ax = nivel_diag.plot(kind='bar', figsize=(10, 6), color=['#59a14f', '#e15759'],
                     edgecolor='white', width=0.7)
plt.title('Diagnostico de Infarto por Nivel Socioeconomico', fontsize=13, fontweight='bold')
plt.xlabel('Nivel Socioeconomico')
plt.ylabel('Cantidad de Pacientes')
plt.xticks(rotation=0)
plt.legend(title='Diagnostico')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=9)

plt.tight_layout()
plt.savefig('../outputs/nivel_socioeconomico_diagnostico.png', dpi=150, bbox_inches='tight')
plt.show()
print(nivel_diag.to_string())
```

## 16. Diagrama de Dispersión — Riesgo Score vs Edad

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Scatter Riesgo_Score vs Edad
colores_diag = df_limpio['Diagnostico_Infarto'].map({0: '#59a14f', 1: '#e15759'})
axes[0].scatter(df_limpio['Edad'], df_limpio['Riesgo_Score'],
                c=colores_diag, alpha=0.4, s=15, edgecolors='none')
axes[0].set_title('Riesgo Score vs Edad por Diagnostico', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Edad (anos)')
axes[0].set_ylabel('Riesgo Score')

from matplotlib.patches import Patch
leyenda = [Patch(facecolor='#59a14f', label='Sin Infarto'),
           Patch(facecolor='#e15759', label='Con Infarto')]
axes[0].legend(handles=leyenda, loc='upper left')

# Scatter Colesterol vs Glucosa
axes[1].scatter(df_limpio['Colesterol_Total'], df_limpio['Glucosa'],
                c=colores_diag, alpha=0.4, s=15, edgecolors='none')
axes[1].set_title('Colesterol Total vs Glucosa', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Colesterol Total (mg/dL)')
axes[1].set_ylabel('Glucosa (mg/dL)')
axes[1].legend(handles=leyenda, loc='upper left')

fig.suptitle('Diagramas de Dispersion — Exploración de Correlaciones', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/scatter_correlaciones.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 17. Mapa de Correlaciones

```python
correlaciones = df_limpio[cols_numericas + ['Diagnostico_Infarto']].corr()

plt.figure(figsize=(12, 9))
mask = np.triu(np.ones_like(correlaciones, dtype=bool))
sns.heatmap(correlaciones, mask=mask, annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, square=True, linewidths=0.5, cbar_kws={'shrink': 0.8},
            annot_kws={'size': 10})
plt.title('Mapa de Correlacion — Variables Numericas del Dataset', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/mapa_correlacion.png', dpi=150, bbox_inches='tight')
plt.show()
print('Correlaciones con Diagnostico_Infarto:')
print(correlaciones['Diagnostico_Infarto'].drop('Diagnostico_Infarto').sort_values(ascending=False).to_string())
```

## 18. Distribución de IMC por Género

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Boxplot IMC por genero
generos = sorted(df_limpio['Genero'].unique())
datos_box = [df_limpio[df_limpio['Genero'] == g]['IMC'].values for g in generos]
colores_box = ['#4e79a7', '#f28e2b']
bp = axes[0].boxplot(datos_box, patch_artist=True,
                     medianprops={'color': 'red', 'linewidth': 2})
axes[0].set_xticklabels(generos)
for patch, color in zip(bp['boxes'], colores_box):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[0].set_title('Distribucion de IMC por Genero', fontsize=12, fontweight='bold')
axes[0].set_ylabel('IMC (kg/m2)')

# Violin plot Edad por diagnostico
for j, (diag, color) in enumerate(zip([0, 1], ['#59a14f', '#e15759'])):
    datos = df_limpio[df_limpio['Diagnostico_Infarto'] == diag]['Edad'].values
    parts = axes[1].violinplot(datos, positions=[j], showmedians=True)
    for pc in parts['bodies']:
        pc.set_facecolor(color)
        pc.set_alpha(0.7)

axes[1].set_title('Distribucion de Edad por Diagnostico de Infarto', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Edad (anos)')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Sin Infarto', 'Con Infarto'])

plt.tight_layout()
plt.savefig('../outputs/imc_edad_distribucion.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 19. Análisis de Factores de Riesgo — Fumador e Historial Familiar

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, col, titulo in zip(axes,
                            ['Fumador', 'Historial_Familiar'],
                            ['Diagnostico segun habito de Fumar',
                             'Diagnostico segun Historial Familiar']):
    tabla = df_limpio.groupby([col, 'Diagnostico_Infarto']).size().unstack(fill_value=0)
    tabla.columns = ['Sin Infarto', 'Con Infarto']
    tabla.plot(kind='bar', ax=ax, color=['#59a14f', '#e15759'], edgecolor='white', width=0.6)
    ax.set_title(titulo, fontsize=12, fontweight='bold')
    ax.set_xlabel(col)
    ax.set_ylabel('Cantidad de Pacientes')
    ax.tick_params(axis='x', rotation=0)
    ax.legend(title='Diagnostico')
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', fontsize=9)

fig.suptitle('Factores de Riesgo vs Diagnostico de Infarto', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/factores_riesgo.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 20. Top 10 Municipios con más Pacientes

```python
top_municipios = df_limpio['Municipio'].value_counts().head(10)

plt.figure(figsize=(14, 6))
colores_mun = sns.color_palette('Blues_r', len(top_municipios))
bars = plt.barh(top_municipios.index[::-1], top_municipios.values[::-1],
                color=colores_mun[::-1], edgecolor='white')

for bar, val in zip(bars, top_municipios.values[::-1]):
    plt.text(bar.get_width() + 5, bar.get_y() + bar.get_height()/2,
             f'{val:,}', va='center', fontsize=10)

plt.title('Top 10 Municipios de Puebla con Mayor Numero de Pacientes', fontsize=13, fontweight='bold')
plt.xlabel('Cantidad de Pacientes')
plt.ylabel('Municipio')
plt.tight_layout()
plt.savefig('../outputs/top_municipios.png', dpi=150, bbox_inches='tight')
plt.show()
print(top_municipios.to_string())
```
