13. Conteo de Pokemones por Generación

Utilizamos **seaborn.countplot()** para crear una gráfica que muestre el conteo de pokemones agrupados por generación. Este gráfico de barras nos permite visualizar la distribución de pokemones a través de las diferentes generaciones. La paleta 'nipy_spectral' proporciona colores distintos para cada barra.

```python
sns.countplot(x='Generation', data=pokemons, palette='nipy_spectral')
plt.title('Number of pokemons grouped by generation')
plt.show()
```

14. Conteo de Pokemones por Tipo

Usamos **.value_counts()** para contar y ordenar los pokemones por tipo. Este método nos devuelve una serie ordenada de mayor a menor frecuencia, mostrándonos cuál es el tipo más común en el dataset y cómo se distribuyen los diferentes tipos de pokemones.

15. Gráfica de Conteo de Tipos de Pokemones

Creamos una visualización más clara del conteo de tipos usando **sns.countplot()** con la paleta 'twilight'. El parámetro figsize=(15,5) establece el tamaño de la figura para mejor legibilidad. Esto nos permite ver de forma visual la distribución de pokemones por tipo.

```python
plt.figure(figsize=(15, 5))
sns.countplot(x='Type', data=pokemons, palette='twilight')
plt.title('Count of Pokemons by Type')
plt.xticks(rotation=45)
plt.show()
```

16. Gráfica de Cajas (Boxplot) para Análisis de Distribución

Un **boxplot** muestra la distribución de datos y detecta valores atípicos (outliers). Usamos **.drop()** con el parámetro `columns=` para eliminar las columnas no numéricas relevantes (ID, Total, Generación, Legendario). La paleta 'seismic' proporciona un contraste visual distintivo. El parámetro fliersize=3 controla el tamaño de los puntos de valores atípicos.

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 6))
sns.boxplot(data=pokemons.drop(columns=['#', 'Name', 'Type', 'Total', 'Generation', 'Legendary']), fliersize=3, palette='seismic')
plt.title('Boxplots for Stats')
plt.show()
```

17. Gráfica de Violín (Violinplot) para Análisis de Distribución

Un **violinplot** es similar al boxplot pero muestra la distribución completa de cada estadística usando densidad de kernel. Es útil para entender la forma de la distribución y comparar múltiples distribuciones. La paleta 'rocket' crea un efecto visual atractivo.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=pokemons.drop(columns=['#', 'Name', 'Type', 'Total', 'Generation', 'Legendary']), palette='rocket')
plt.title('Violinplots for Stats')
plt.show()
```

18. Agrupación por Tipo y Suma de Estadísticas

Usamos **.groupby('Type').sum()** para agrupar pokemones por tipo y sumar todas las estadísticas numéricas para cada tipo. Esto nos permite comparar el valor total de cada estadística por tipo de pokemón, dándonos información agregada sobre las características de cada tipo.

```python
pokemons.groupby('Type').sum()
```

19. Extracción de un Atributo Específico Agrupado

Accedemos a un atributo específico del resultado del groupby usando notación de punto. En este caso, obtenemos la suma total de HP (puntos de salud) para cada tipo de pokemón. Los tipos aparecen en orden alfabético en la serie resultante, permitiéndonos identificar qué tipo tiene la resistencia total más alta.

```python
pokemons.groupby('Type').sum()['HP']

# In this series the types are alphabetically ordered.
# Muestra el HP total por tipo en orden alfabético
```
