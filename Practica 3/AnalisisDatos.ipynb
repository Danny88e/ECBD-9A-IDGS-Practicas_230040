## Tabla de Frecuencia: Distribución de Estadísticas por Tipo

Visualización detallada de cada estadística para cada tipo de pokemón usando strip plot:

```python
# Preparar datos para la tabla de frecuencia
# Crear un DataFrame con las estadísticas relevantes
stats_data = pokemons[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Crear la visualización de tabla de frecuencia con strip plot
plt.figure(figsize=(15, 30))

for k, stat in enumerate(stats_data.columns, start=1):
    plt.subplot(6, 1, k)
    
    sns.stripplot(
        data=pokemons,
        x='Type',
        y=stat,
        hue='Type',
        palette='Dark2',
        legend=False,
        jitter=True,
        size=6
    )
    
    plt.title(f'{k}. Distribución de {stat} por Tipo', fontsize=12, fontweight='bold')
    plt.xlabel('Tipo')
    plt.ylabel(stat)
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

23. Visualización de Strip Plot: Distribución de Estadísticas por Tipo

Un **strip plot** es una técnica de visualización que muestra todos los puntos individuales de datos distribuidos aleatoriamente a lo largo del eje X para evitar superposición. Este gráfico es especialmente útil para ver la densidad y distribución de valores de cada estadística (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) para cada tipo de pokemón. La paleta 'Dark2' proporciona colores contrastantes para una fácil identificación visual.

```python
# Usar solo las 6 estadísticas principales
stats_principales = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

k = 1
plt.figure(figsize=(15, 30))
for i in stats_principales:
    plt.subplot(6, 1, k)
    k = k + 1
    sns.stripplot(x=pokemons.Type, y=pokemons[i], hue=pokemons.Type, palette='Dark2', legend=False)
    plt.title(f'Total of {i} for each type', fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

```

25. Transformación de Datos: Pivoting del Dataset con Melt

La función **pd.melt()** transforma datos de formato "wide" a formato "long". En este caso, tomamos el dataset original y convertimos todas las estadísticas (HP, Defense, Attack, Sp. Atk, Sp. Def, Speed) en una única columna de "variable" con sus correspondientes "values". Mantenemos como identificadores (id_vars) el Name, Type y Legendary del pokemón. Esta transformación es útil para crear visualizaciones más complejas y realizar análisis que requieren datos en formato largo.

```python
# Transformación del dataset original a formato largo (melt)
pok_melt = pd.melt(pokemons, 
                    id_vars=['Name', 'Type', 'Legendary'],
                    value_vars=['HP', 'Defense', 'Attack', 'Sp. Atk', 'Sp. Def', 'Speed'])


print("\n📊 TABLA TRANSFORMADA - Primeros 20 registros:")
print(pok_melt.head(20))
```

26. Análisis de Distribución Individual por Tipo usando Swarmplot

Un **swarmplot** muestra todos los puntos de datos individuales distribuidos para cada variable (estadística), evitando superposiciones. Este gráfico es excelente para visualizar la distribución completa de valores de HP, Attack, Defense, Sp. Atk, Sp. Def y Speed para cada tipo específico de pokemón. Creamos una cuadrícula de 7x3 subplots (uno para cada tipo) con la paleta 'gist_stern' para una mejor diferenciación visual. Cada subplot muestra cómo se distribuyen las 6 estadísticas principales dentro de cada tipo.

```python
# Crear un DataFrame con el promedio de cada estadística por tipo
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df = pd.DataFrame()
for i in stats:
    df[i] = pokemons.groupby('Type').describe()[i]['mean']

print("📊 TABLA DE PROMEDIOS DE ESTADÍSTICAS POR TIPO:")
print(df)
```

29. Visualización Comparativa de Promedios por Estadística
